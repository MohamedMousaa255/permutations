q2) part2
###backtracking based recursive solution###

```python
def permutation(list_):
    if len(list_) == 1:
        return [list_]
    if len(list_) == 0:
        return []
    l1 = [] 
    for i in range(len(list_)):
        n = list_[i]
        remLst = list_[:i] + list_[i+1:]
        for p in permutation(remLst):
            l1.append([n] + p)
    return l1
```

```python
num = list('34789')
for x in permutation(num):
    print (x)
```
